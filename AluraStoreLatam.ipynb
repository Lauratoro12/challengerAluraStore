### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
print(f"Ingreso total Tienda 1: ${ingreso_tienda1}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda2}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda3}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
# Tienda 1
ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Tienda 1 - Productos vendidos por categoría:")
print(ventas_categoria_tienda1)
# Tienda 2
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("\nTienda 2 - Productos vendidos por categoría:")
print(ventas_categoria_tienda2)

# Tienda 3
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("\nTienda 3 - Productos vendidos por categoría:")
print(ventas_categoria_tienda3)

# Tienda 4
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("\nTienda 4 - Productos vendidos por categoría:")
print(ventas_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio por tienda
calificacion_prom_tienda1 = tienda['Calificación'].mean()
calificacion_prom_tienda2 = tienda2['Calificación'].mean()
calificacion_prom_tienda3 = tienda3['Calificación'].mean()
calificacion_prom_tienda4 = tienda4['Calificación'].mean()

# Mostrar resultados
print(f"Calificación promedio Tienda 1: {calificacion_prom_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_prom_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_prom_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_prom_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Tienda 1
ventas_productos_tienda1 = tienda.groupby('Producto')['Precio'].count().sort_values(ascending=False)
producto_mas_vendido_tienda1 = ventas_productos_tienda1.idxmax()
producto_menos_vendido_tienda1 = ventas_productos_tienda1.idxmin()

print("Tienda 1")
print("Producto más vendido:", producto_mas_vendido_tienda1)
print("Producto menos vendido:", producto_menos_vendido_tienda1)
print(ventas_productos_tienda1.head(5))  # Top 5 productos
print(ventas_productos_tienda1.tail(5))  # 5 menos vendidos

# Tienda 2
ventas_productos_tienda2 = tienda2.groupby('Producto')['Precio'].count().sort_values(ascending=False)
producto_mas_vendido_tienda2 = ventas_productos_tienda2.idxmax()
producto_menos_vendido_tienda2 = ventas_productos_tienda2.idxmin()

print("\nTienda 2")
print("Producto más vendido:", producto_mas_vendido_tienda2)
print("Producto menos vendido:", producto_menos_vendido_tienda2)
print(ventas_productos_tienda2.head(5))
print(ventas_productos_tienda2.tail(5))

# Tienda 3
ventas_productos_tienda3 = tienda3.groupby('Producto')['Precio'].count().sort_values(ascending=False)
producto_mas_vendido_tienda3 = ventas_productos_tienda3.idxmax()
producto_menos_vendido_tienda3 = ventas_productos_tienda3.idxmin()

print("\nTienda 3")
print("Producto más vendido:", producto_mas_vendido_tienda3)
print("Producto menos vendido:", producto_menos_vendido_tienda3)
print(ventas_productos_tienda3.head(5))
print(ventas_productos_tienda3.tail(5))

# Tienda 4
ventas_productos_tienda4 = tienda4.groupby('Producto')['Precio'].count().sort_values(ascending=False)
producto_mas_vendido_tienda4 = ventas_productos_tienda4.idxmax()
producto_menos_vendido_tienda4 = ventas_productos_tienda4.idxmin()

print("\nTienda 4")
print("Producto más vendido:", producto_mas_vendido_tienda4)
print("Producto menos vendido:", producto_menos_vendido_tienda4)
print(ventas_productos_tienda4.head(5))
print(ventas_productos_tienda4.tail(5))
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio por tienda
costo_envio_prom_tienda1 = tienda['Costo de envío'].mean()
costo_envio_prom_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_prom_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_prom_tienda4 = tienda4['Costo de envío'].mean()

# Mostrar resultados
print(f"Costo de envío promedio Tienda 1: ${costo_envio_prom_tienda1:.2f}")
print(f"Costo de envío promedio Tienda 2: ${costo_envio_prom_tienda2:.2f}")
print(f"Costo de envío promedio Tienda 3: ${costo_envio_prom_tienda3:.2f}")
print(f"Costo de envío promedio Tienda 4: ${costo_envio_prom_tienda4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Datos de ingresos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos_totales = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos_totales, color='skyblue')
plt.title("Ingreso Total por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso Total ($)")
plt.show()
```

```python
# Ventas por categoría para Tienda 1
ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Producto'].count()

plt.figure(figsize=(7,7))
plt.pie(ventas_categoria_tienda1, labels=ventas_categoria_tienda1.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribución de Productos Vendidos por Categoría - Tienda 1")
plt.show()
```

```python
# Top 5 productos más vendidos en Tienda 1
top5_productos_tienda1 = tienda.groupby('Producto')['Precio'].count().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
top5_productos_tienda1.plot(kind='barh', color='orange')
plt.title("Top 5 Productos Más Vendidos - Tienda 1")
plt.xlabel("Cantidad Vendida")
plt.gca().invert_yaxis()  # Mostrar el más vendido arriba
plt.show()
```
